## Q1. Python function — F-value and p-value for variance ratio test

```python
import numpy as np
from scipy import stats

def variance_ratio_test(arr1, arr2):
    """
    Perform F-test for equality of variances (two-tailed).
    Parameters:
        arr1, arr2 : array-like data samples
    Returns:
        F-value, p-value
    """
    var1 = np.var(arr1, ddof=1)
    var2 = np.var(arr2, ddof=1)

    # Put larger variance in numerator
    if var1 >= var2:
        F = var1 / var2
        df1 = len(arr1) - 1
        df2 = len(arr2) - 1
    else:
        F = var2 / var1
        df1 = len(arr2) - 1
        df2 = len(arr1) - 1

    # Two-tailed p-value
    p_value = 2 * min(stats.f.cdf(F, df1, df2),
                      stats.f.sf(F, df1, df2))

    return F, p_value

# Example
np.random.seed(42)
a = np.random.normal(0, 1, 20)
b = np.random.normal(0, 1.5, 25)

F, p = variance_ratio_test(a, b)
print(f"F-value : {F:.4f}")
print(f"p-value : {p:.4f}")
```

## Q2. Python function — Critical F-value for two-tailed test

```python
from scipy import stats

def critical_f_value(df_num, df_den, alpha=0.05):
    """
    Returns the critical F-value for a two-tailed F-test.
    Parameters:
        df_num : degrees of freedom for numerator
        df_den : degrees of freedom for denominator
        alpha  : significance level (default 0.05)
    Returns:
        critical F-value (upper tail only, since F is always positive)
    """
    # Two-tailed: alpha/2 in each tail
    f_critical = stats.f.ppf(1 - alpha/2, df_num, df_den)
    return f_critical

# Example
df_num = 5
df_den = 10
alpha  = 0.05

f_crit = critical_f_value(df_num, df_den, alpha)
print(f"df_numerator   : {df_num}")
print(f"df_denominator : {df_den}")
print(f"Alpha          : {alpha}")
print(f"Critical F-value (two-tailed, upper): {f_crit:.4f}")
print(f"\nReject H0 if F > {f_crit:.4f}")
```

## Q3. F-test program — Random samples from two normal distributions

```python
import numpy as np
from scipy import stats

np.random.seed(42)

# Known variances
var1, var2 = 4, 9    # sigma^2
n1,   n2   = 30, 30

# Generate samples
sample1 = np.random.normal(0, np.sqrt(var1), n1)
sample2 = np.random.normal(0, np.sqrt(var2), n2)

s1_var = np.var(sample1, ddof=1)
s2_var = np.var(sample2, ddof=1)

# F-statistic (larger / smaller)
if s1_var >= s2_var:
    F    = s1_var / s2_var
    df1  = n1 - 1
    df2  = n2 - 1
else:
    F    = s2_var / s1_var
    df1  = n2 - 1
    df2  = n1 - 1

p_value = 2 * min(stats.f.cdf(F, df1, df2), stats.f.sf(F, df1, df2))

alpha = 0.05
print("===== F-Test: Two Normal Distributions =====")
print(f"Population variance 1 (known) : {var1}")
print(f"Population variance 2 (known) : {var2}")
print(f"Sample variance 1             : {s1_var:.4f}")
print(f"Sample variance 2             : {s2_var:.4f}")
print(f"\nF-statistic           : {F:.4f}")
print(f"df1 (numerator)       : {df1}")
print(f"df2 (denominator)     : {df2}")
print(f"p-value               : {p_value:.4f}")

if p_value < alpha:
    print("\nResult: REJECT H₀ — Variances are significantly different.")
else:
    print("\nResult: FAIL TO REJECT H₀ — Variances are NOT significantly different.")
```

## Q4. F-test — Population variances 10 and 15, n=12 each, α=0.05

```python
from scipy import stats

# Given
var1 = 10
var2 = 15
n1   = 12
n2   = 12
alpha = 0.05

# F-statistic = larger variance / smaller variance
F   = var2 / var1    # 15/10 = 1.5
df1 = n2 - 1         # 11
df2 = n1 - 1         # 11

# Two-tailed p-value
p_value  = 2 * stats.f.sf(F, df1, df2)
f_critical = stats.f.ppf(1 - alpha/2, df1, df2)

print("===== F-Test: Population Variances 10 vs 15 =====")
print(f"Population Variance 1 : {var1}")
print(f"Population Variance 2 : {var2}")
print(f"n1 = n2               : {n1}")
print(f"F-statistic           : {F:.4f}")
print(f"df1, df2              : {df1}, {df2}")
print(f"Critical F (α=0.05)   : {f_critical:.4f}")
print(f"p-value               : {p_value:.4f}")

if p_value < alpha:
    print("\nResult: REJECT H₀ — Variances are significantly different at 5% level.")
else:
    print("\nResult: FAIL TO REJECT H₀ — Variances are NOT significantly different at 5% level.")
```

## Q5. F-test — Manufacturer variance claim 0.005, sample var=0.006, n=25, α=0.01

```python
from scipy import stats

# Given
claimed_var = 0.005
sample_var  = 0.006
n           = 25
alpha       = 0.01

# F-statistic = sample_var / claimed_var
F   = sample_var / claimed_var
df1 = n - 1     # 24
df2 = n - 1     # 24 (reference)

# Two-tailed p-value
p_value    = 2 * stats.f.sf(F, df1, df2)
f_critical = stats.f.ppf(1 - alpha/2, df1, df2)

print("===== F-Test: Manufacturer Variance Claim =====")
print(f"Claimed variance  : {claimed_var}")
print(f"Sample variance   : {sample_var}")
print(f"n                 : {n}")
print(f"F-statistic       : {F:.4f}")
print(f"df1, df2          : {df1}, {df2}")
print(f"Critical F (α=0.01): {f_critical:.4f}")
print(f"p-value           : {p_value:.4f}")

if p_value < alpha:
    print("\nResult: REJECT H₀ — Claim is NOT justified. Variance significantly differs from 0.005.")
else:
    print("\nResult: FAIL TO REJECT H₀ — Claim IS justified. No significant difference from 0.005.")
```

## Q6. Python function — Mean and Variance of F-distribution

```python
def f_distribution_mean_variance(df_num, df_den):
    """
    Calculate mean and variance of an F-distribution.
    Parameters:
        df_num : degrees of freedom numerator (d1)
        df_den : degrees of freedom denominator (d2)
    Returns:
        (mean, variance) as a tuple
    """
    if df_den <= 2:
        mean = float('inf')   # Mean undefined for df_den <= 2
    else:
        mean = df_den / (df_den - 2)

    if df_den <= 4:
        variance = float('inf')   # Variance undefined for df_den <= 4
    else:
        variance = (2 * df_den**2 * (df_num + df_den - 2)) / \
                   (df_num * (df_den - 2)**2 * (df_den - 4))

    return mean, variance

# Example
df_num = 5
df_den = 10

mean, variance = f_distribution_mean_variance(df_num, df_den)
print(f"F-distribution with df_num={df_num}, df_den={df_den}")
print(f"Mean     : {mean:.4f}")
print(f"Variance : {variance:.4f}")

# Test multiple values
print("\nDifferent df combinations:")
for d1, d2 in [(3, 8), (5, 20), (10, 30)]:
    m, v = f_distribution_mean_variance(d1, d2)
    print(f"  df1={d1}, df2={d2} → mean={m:.4f}, variance={v:.4f}")
```

## Q7. F-test — Sample variances 25 (n=10) vs 20 (n=15), α=0.10

```python
from scipy import stats

# Given
s1_var, n1 = 25, 10
s2_var, n2 = 20, 15
alpha      = 0.10

# F = larger / smaller
F   = s1_var / s2_var   # 25/20 = 1.25
df1 = n1 - 1            # 9
df2 = n2 - 1            # 14

# Two-tailed p-value
p_value    = 2 * stats.f.sf(F, df1, df2)
f_critical = stats.f.ppf(1 - alpha/2, df1, df2)

print("===== F-Test: Unknown Variances (α=0.10) =====")
print(f"Sample 1 → variance={s1_var}, n={n1}, df={df1}")
print(f"Sample 2 → variance={s2_var}, n={n2}, df={df2}")
print(f"F-statistic        : {F:.4f}")
print(f"Critical F (α=0.10): {f_critical:.4f}")
print(f"p-value            : {p_value:.4f}")

if p_value < alpha:
    print("\nResult: REJECT H₀ — Variances are significantly different at 10% level.")
else:
    print("\nResult: FAIL TO REJECT H₀ — Variances are NOT significantly different at 10% level.")
```

## Q8. F-test — Restaurant waiting times, α=0.05

```python
import numpy as np
from scipy import stats

# Given data
rest_A = np.array([24, 25, 28, 23, 22, 20, 27])
rest_B = np.array([31, 33, 35, 30, 32, 36])

var_A = np.var(rest_A, ddof=1)
var_B = np.var(rest_B, ddof=1)
n_A   = len(rest_A)
n_B   = len(rest_B)

# F = larger / smaller
if var_A >= var_B:
    F = var_A / var_B;  df1 = n_A - 1;  df2 = n_B - 1
else:
    F = var_B / var_A;  df1 = n_B - 1;  df2 = n_A - 1

alpha      = 0.05
p_value    = 2 * min(stats.f.cdf(F, df1, df2), stats.f.sf(F, df1, df2))
f_critical = stats.f.ppf(1 - alpha/2, df1, df2)

print("===== F-Test: Restaurant Waiting Times =====")
print(f"Restaurant A → data={rest_A.tolist()}, n={n_A}")
print(f"Restaurant B → data={rest_B.tolist()}, n={n_B}")
print(f"\nVariance A : {var_A:.4f}")
print(f"Variance B : {var_B:.4f}")
print(f"F-statistic        : {F:.4f}")
print(f"df1, df2           : {df1}, {df2}")
print(f"Critical F (α=0.05): {f_critical:.4f}")
print(f"p-value            : {p_value:.4f}")

if p_value < alpha:
    print("\nResult: REJECT H₀ — Variances are significantly different at 5% level.")
else:
    print("\nResult: FAIL TO REJECT H₀ — Variances are NOT significantly different at 5% level.")
```

## Q9. F-test — Student test scores Group A vs Group B, α=0.01
